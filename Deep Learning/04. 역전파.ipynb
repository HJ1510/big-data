3) 교차 엔트로피(Cross Entropy)

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(start=0.00001, stop=1.0, num=100)
plt.plot(x, -np.log(x), 'b-', label='Natural Logarithm(ln)')
plt.legend()
plt.show()
```

정답에 가까울 수록 전체 오차 값은 작아지며, 반대로 정답에서 멀어질수록 한없이 커짐

y가 1에 가까울수록 오차가 적다

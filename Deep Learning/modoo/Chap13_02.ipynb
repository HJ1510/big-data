```python
import pandas as pd
df = pd.read_csv('data/sonar3.csv', header=None)
```

```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import TensorBoard
from time import time
```

```python
X = df.iloc[:, :60]
y = df.iloc[:, 60]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
model = models.Sequential()
model.add(layers.Dense(24, input_dim=60, activation='relu'))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
tensorboard = TensorBoard(log_dir='logs\{}'.format(time()))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=200, batch_size=10, callbacks=[tensorboard])
```

터미널에서 logs 폴더 존재하는 디렉토리로 이동하여  
tensorboard --logdir=.\logs  
치면 tensorboard 사용 가능

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy: ', score[1]) #score[0] 손실값, score[1] 정확도
print('Test Score: ', score[0]) #score[0] 손실값, score[1] 정확도
```

4. 학습의 자동 중단

```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os
import pandas as pd
```

```python
df = pd.read_csv('data/wine.csv', header=None)
X = df.iloc[:, :12]
y = df.iloc[:, 12]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
model = models.Sequential()
model.add(layers.Dense(30, input_dim=12, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=20) 
# monitor='val_loss' val_loss를 모니터 하다가 
# patience=20 연속적으로 20번 이상 좋아지지 않으면(val_loss가 줄어들지 않으면)
# 강제로 학습 종료
model_path = './model/best_model.hdf5'
checkpoint = ModelCheckpoint(filepath=model_path, monitor='val_loss', verbose=0,
                            save_best_only=True)
# monitor='val_loss' val_loss를 모니터 하다가 
# save_best_only 이전에 저장했던것보다 더 좋아졌을 때만(val_loss가 줄어들었을 때만) 저장
```

```python
history = model.fit(X_train, y_train, epochs=2000, batch_size=500, validation_split=0.25, verbose=1,
                   callbacks=[early_stopping, checkpoint])
```

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy:', score[1])
```

```python
from urllib.parse import urljoin
from tqdm.notebook import tqdm
```

```python
base_url = 'https://www.chicagomag.com'  #해당 주소가 안 붙여진 값 존재

rank = []
sandwich_name = []
cafe_name = []
link_url = []

list_soup = soup.select('div.sammy')

for info in tqdm(list_soup):
    rank.append(info.select('.sammyRank')[0].text)
    list_string = info.select('.sammyListing')[0].text.split('\n')
    sandwich_name.append(list_string[0])
    cafe_name.append(list_string[1])
    link_url.append(urljoin(base_url, info.find('a')['href']))
```

```python
link_url
```

```python
import pandas as pd
```

```python
data = {'Rank':rank, 'Sandwich':sandwich_name, 'Cafe': cafe_name, 'URL': link_url}
df = pd.DataFrame(data)
df.head()
```

```python
import time
```

```python
price = []
address = []

for index in tqdm(df.index):
    html = requests.get(df['URL'][index], headers={"User-Agent":"Mozilla/5.0"})
    time.sleep(1)
    tmp = BeautifulSoup(html.text, 'lxml')
    val_list = tmp.select('p.addy')[0].text.split(',')[0].split()
    price.append(val_list[0][:-1])
    address.append(' '.join(val_list[1:]))
```

```python
print(len(price), len(address))
```

```python
address
```

```python
df['Price'] = price
df['Address'] = address
df.head()
```

```python
df.set_index('Rank', inplace=True)
df.head()
```

```python
import folium
import googlemaps
import numpy as np
```

```python
gmaps_key = "*****"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm(df.index):
    if df['Address'][n] != 'Multiple locations':
        target_name = df['Address'][n] + ', ' + 'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        time.sleep(0.5)
        location = gmaps_output[0].get('geometry')
        lat.append(location['location']['lat'])
        lng.append(location['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
print(len(lat), len(lng))
```

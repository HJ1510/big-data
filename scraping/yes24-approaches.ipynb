함수 생성 실제 크롤링때 사용한 함수

```python
def yes24DataReader(CategoryNum, startPage, endPage):

    root_url = 'http://www.yes24.com'

    url_1 = 'http://www.yes24.com/24/Category/Display/'
    url_2 = '?PageNumber='
    url_set = url_1 + CategoryNum + url_2
    
    book_url_list=[]
    book_list=[]
    for i in range(startPage, endPage+1):

        origin_books_tag = 'div.cCont_goodsSet'
        url = url_set + str(i)
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'lxml')        
        origin_books = soup.select(origin_books_tag)

        # 수집 중인 페이지 번호 출력
        print('# Page', i)

        # 개별 도서 정보 수집
        for origin_book in origin_books:
            origin_book_img_url = origin_book.select_one('p img').get('src')
            if origin_book_img_url != "http://image.yes24.com/sysimage/pd_19_L.gif": # 성인인증여부를 img로 확인
                # 책 정보에 접근 가능한 URL 추출
                url_tag = 'a.bgYUI.ico_nWin'
                book = origin_book.select_one(url_tag)                
                book_url = book.get('href')
                book_url_list.append(book_url)
                sub_url = root_url + book_url
                sub_html = requests.get(sub_url).text
                sub_soup = BeautifulSoup(sub_html, 'lxml')
                info_tag = 'strong.txC_blue' # yes24 구성상품 확인
                book_info = sub_soup.select(info_tag)[0].text
                
                if "구성한" in book_info:
#                     print(book_info)
                    continue
                else:
                    title_tag = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2'
                    author_tag = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth > a' 
                    author_tag2 = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth'
                    author_tag3 = 'div.authorTit > div.author_name'
                    publisher_tag = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_pub > a'
                    isbn_tag = '#infoset_specific > div.infoSetCont_wrap > div > table > tbody > tr:nth-child(3) > td'
                    img_tag = 'img.gImg'
                    category_tag = '#infoset_goodsCate > div.infoSetCont_wrap > dl > dd > ul > li:nth-child(1) > a:nth-child(4)'

                    book_title = sub_soup.select(title_tag)[0].text.strip()


                    try: # 저자 예외 처리
                        book_author = sub_soup.select(author_tag)[0].text
                    except IndexError:
                        try:
                            book_author = sub_soup.select(author_tag2)[0].text.replace("\r\n", "").replace(" ","")
                        except IndexError:
                            try:
                                book_author_text = sub_soup.select(author_tag3)[0].text
                                if "저자 :" in book_author_text:
                                    book_author = book_author_text.split("저자 :")[1].strip()
                                else:
                                    book_author = book_author_text.strip()      
                            except IndexError:
                                book_author = "저자 정보없음"
                    try :
                        book_publisher = sub_soup.select(publisher_tag)[0].text
                    except :
                        book_publisher = "출판사 정보없음"

                    try :
                        book_isbn = sub_soup.find('th', text='ISBN13').find_next_sibling('td').text
                    except :
                        book_isbn = ''

                    book_img = sub_soup.select(img_tag)[0].get('src')

                    try :
                        category_mapping = {} # csv 파일 이용해서 매핑
                        with open('./book_category_202303091301.csv', newline='', encoding='utf-8') as csvfile:
                            reader = csv.DictReader(csvfile)
                            for row in reader:
                                category_mapping[row['category']] = int(row['id']) # 카테고리 매핑 정보를 딕셔너리에 저장

                        book_category = sub_soup.select(category_tag)[0].text # 웹에서 수집한 카테고리 정보 소설/시/희곡

                        if book_category in category_mapping: # 매핑된 카테고리가 있다면
                            book_category_id = category_mapping[book_category] # 카테고리를 숫자로 매핑
                        else:
                            book_category_id = 27 # 매핑된 카테고리가 없다면 None으로 처리
                    except :
                        book_category_id = 27

                    book_list.append([book_title, book_author, book_publisher, book_isbn, book_img, book_category_id])

#                     print('=========>', book_title)                  
#                     print('=========>', book_publisher)
            else:
                continue


    # 데이터프레임 컬럼명 지정
    colList = ['book_title',  'book_author', 'book_publisher',  'book_ISBN', 'book_img_url', 'book_category_id']
    print(len(book_list))
    # 데이터프레임으로 변환
    df = pd.DataFrame(book_list, columns=colList)

    return df
```

```python
now = datetime.datetime.now()
current_time = now.strftime("%Y-%m-%d_%H-%M-%S")

CategoryNumber='001001046004' #소설/시/희곡>중국소설

startPage=1
endPage=102

# 수집한 데이터를 저장할 빈 리스트
data_list = []

# 크롤링 시작
for page in range(startPage, endPage+1):
    try:
        # 페이지 이동 및 데이터 수집
        df = yes24DataReader(CategoryNumber, page, page)
        
        # 수집한 데이터를 리스트에 추가
        data_list.append(df)
        
        # 로그 출력
#         print(f"Page {page} data has been collected.")
        
    except Exception as e:
        # 오류가 발생한 경우 로그 출력
        print(f"Error occurred onE:\nodeStudy\expressPractice\cowork\app.js page {page}: {e}")
        
        # 수집한 데이터를 csv 파일로 저장
        if data_list:
            df_all = pd.concat(data_list)
            df_all.to_csv(str(current_time)+'_'+'소설시희곡_중국소설'+'_'+str(startPage)+' to '+str(endPage)+'.csv', index=False, encoding='utf-8')
        
        # 오류가 발생한 페이지 이후의 데이터 수집 중단
        break

# 모든 데이터를 csv 파일로 저장
if data_list:
    df_all = pd.concat(data_list)
    df_all.to_csv(str(current_time)+'_'+'소설시희곡_중국소설'+'_'+str(startPage)+' to '+str(endPage)+'.csv', index=False, encoding='utf-8')
```

개별 페이지 크롤링 확인 (함수 생성 실제 크롤링때 사용한 함수)

```python
url ='http://www.yes24.com/Product/Goods/102792311'
html = requests.get(url).text
soup = BeautifulSoup(html, 'lxml')
title_tag = 'h2.gd_name'
author_tag = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth > a' 
author_tag2 = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth'
author_tag3 = 'div.authorTit > div.author_name'
publisher_tag = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_pub > a'
isbn_tag = '#infoset_specific > div.infoSetCont_wrap > div > table > tbody > tr:nth-child(3) > td'
img_tag = 'img.gImg'
category_tag = '#infoset_goodsCate > div.infoSetCont_wrap > dl > dd > ul > li:nth-child(1) > a:nth-child(4)'

info_tag = 'strong.txC_blue' # yes24 구성상품 확인
                

book_title = soup.select(title_tag)[0].text.strip()


try: # 저자 예외 처리
    book_author = soup.select(author_tag)[0].text
except IndexError:
    try:
        book_author = soup.select(author_tag2)[0].text.replace("\r\n", "").replace(" ","")
    except IndexError:
        try:
            book_author_text = soup.select(author_tag3)[0].text
            if "저자 :" in book_author_text:
                book_author = book_author_text.split("저자 :")[1].strip()
            else:
                book_author = book_author_text.strip()  
        except IndexError:
            book_author = "저자 정보없음"
            
try :
    book_publisher = soup.select(publisher_tag)[0].text
except :
    book_publisher = "출판사 정보없음"

try :
    book_isbn = soup.find('th', text='ISBN13').find_next_sibling('td').text
except :
    book_isbn = ''
    
book_info = soup.select(info_tag)[0].text

book_img = soup.select(img_tag)[0].get('src')

category_mapping = {} # csv 파일 이용해서 매핑
with open('./book_category_202303091301.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        category_mapping[row['category']] = int(row['id']) # 카테고리 매핑 정보를 딕셔너리에 저장

book_category = soup.select(category_tag)[0].text # 웹에서 수집한 카테고리 정보 소설/시/희곡

if book_category in category_mapping: # 매핑된 카테고리가 있다면
    book_category_id = category_mapping[book_category] # 카테고리를 숫자로 매핑
else:
    book_category_id = 27 # 매핑된 카테고리가 없다면 None으로 처리


print(book_title)
print(book_author)
print(book_publisher)
print(book_isbn)
print(book_img)
print(book_category)
print(book_category_id)
print(book_info)
```
